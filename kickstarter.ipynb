```python
import sklearn 
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
fn = '/Users/crazydonkey/Downloads/funding-successful-projects-on-kickstarter/train.csv'
f = open(fn)
df = pd.read_csv(f)

fn = '/Users/crazydonkey/Downloads/funding-successful-projects-on-kickstarter/test.csv'
f = open(fn)
test_df = pd.read_csv(f)
```

```python
df
```

```python
def encoder_cat(df, col):
    SuccessList = df[col].tolist()
    labelEncoderSuccess = LabelEncoder()
    labelEncoderSuccess.fit(SuccessList)
    labelsSuccess = labelEncoderSuccess.transform(SuccessList)
    df[col]=pd.Series(labelsSuccess)
    
def preprocess(df):
    df = df.drop(columns=['desc', 'keywords', 'name', 'project_id'])
    encoder_cat(df, 'country')
    encoder_cat(df, 'currency')
    encoder_cat(df, 'disable_communication')
    return df
```

```python
df = preprocess(df)
df = df.sample(frac=1)
```

```python
data = df.values
```

```python
n = float(len(data))
n_train = int(n * 0.8)
train_data = data[:n_train,:]
test_data = data[n_train:,:]
```
