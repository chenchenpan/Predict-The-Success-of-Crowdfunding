```python
%matplotlib inline
import numpy as np
import pandas as pd
import json
pd.set_option('display.max_colwidth', -1)
```

```python
# load Kickstarter data
data = [json.loads(line) for line in open('Kickstarter_2019-12-12T03_20_05_306Z.json', 'r')]
```

```python
data[0]['data']
```

```python
d = {k: data[0]['data'][k] for k in ('id','name','blurb','goal',
                                     'pledged',
                                     'state',
                                     'deadline',
                                     'state_changed_at',
                                     'created_at',
                                     'launched_at',
                                     'country',
                                     'currency',
                                     'backers_count'
                                     )}
d['creator_id'] = data[0]['data']['creator']['id']
d['project_id'] = data[0]['data']['profile']['project_id']
d['category_id'] = data[0]['data']['category']['id']
d['category_name'] = data[0]['data']['category']['name']
d['category_parent_id'] = data[0]['data']['category']['parent_id']
d['category_slug'] = data[0]['data']['category']['slug']
```

```python
d
```

```python
# extract meaningful features and to create dataframe
def extract_features(data):
    data_dict = {}
    for i, exam in enumerate(data):
        d = {k: data[i]['data'][k] for k in ('id','name','blurb','goal',
                                             'pledged',
                                             'state',
                                             'deadline',
                                             'state_changed_at',
                                             'created_at',
                                             'launched_at',
                                             'country',
                                             'currency',
                                             'backers_count'
                                             )}
        d['creator_id'] = data[i]['data']['creator']['id']
        d['project_id'] = data[i]['data']['profile']['project_id']
        d['category_id'] = data[i]['data']['category']['id']
        d['category_name'] = data[i]['data']['category']['name']
#         d['category_parent_id'] = data[i]['data']['category']['parent_id']
        d['category_slug'] = data[i]['data']['category']['slug']
        
        data_dict[i] = d
    return data_dict
    
```

```python
dic = extract_features(data)
```

```python
len(dic.keys())
```

```python
df = pd.DataFrame.from_dict(dic, orient='index')
```

```python
df.head()
```

```python
df.shape
```

```python
df.to_csv('kickstarter-2019-extracted.csv', index=None)
```
